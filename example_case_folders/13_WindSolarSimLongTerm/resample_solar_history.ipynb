## Reload data to test

```python
data = pd.read_csv("inputs/Flatirons_solar_data_sunset_1s.csv")
data["Timestamp"] = pd.DatetimeIndex(pd.to_datetime(data["Timestamp"], format="ISO8601"))
data = data.set_index("Timestamp")
data.head()
```

```python
# make dictionary for column names in case they get moved around
col_dict = {}
for i, col in enumerate(data.columns):
    if "Global Horizontal Irradiance" in col:
        col_dict["ghi_col"] = data.columns.get_loc(col) + 1  # bc 1st col will be timestamp
    elif "Direct Normal Irradiance" in col:
        col_dict["dni_col"] = data.columns.get_loc(col) + 1  # bc 1st col will be timestamp
    elif "Diffuse Horizontal Irradiance" in col:
        col_dict["dhi_col"] = data.columns.get_loc(col) + 1  # bc 1st col will be timestamp
    elif "Temperature" in col:
        col_dict["temp_col"] = data.columns.get_loc(col) + 1  # bc 1st col will be timestamp
    elif "Wind Speed at 19" in col:
        col_dict["ws_col"] = data.columns.get_loc(col) + 1  # bc 1st col will be timestamp

col_dict
```

```python
weather_data_array = data.reset_index().to_numpy()
weather_data_array
```

```python
weather_data_array[0, 0]
```

```python
time_index = weather_data_array[0, 0] + pd.Timedelta(seconds=2.0)
# time_index

condition = weather_data_array[:, 0] == time_index
row_index = np.where(condition)[0][0]
weather_data_array[row_index, :]
```

```python
row_index
```

```python
col_dict["dni_col"]
```

```python
weather_data = np.array(
    [
        [time_index.year],
        [time_index.month],
        [time_index.day],
        [time_index.hour],
        [time_index.minute],
        [weather_data_array[row_index, col_dict["dni_col"]]],
        [weather_data_array[row_index, col_dict["dhi_col"]]],
        [weather_data_array[row_index, col_dict["ghi_col"]]],
        [weather_data_array[row_index, col_dict["ws_col"]]],
        [weather_data_array[row_index, col_dict["temp_col"]]],
    ]
)

weather_data
```
